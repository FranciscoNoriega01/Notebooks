```
%pythonimport time
from datetime import datetime, timedelta

# Set the number of minutes to run
run_minutes = 70  # desired duration
end_time = datetime.now() + timedelta(minutes=run_minutes)  # Calculate end time

# Loop until the specified time
while datetime.now() < end_time:
    # Print the current time
    print("Current time:", datetime.now())
    
    # 1 second pause to make the output readable
    time.sleep(1)
```
